На этом этапе мы займемся очисткой, преобразованием и подготовкой данных для дальнейшего анализа и моделирования. Это включает обработку пропусков, корректировку типов данных, устранение аномальных значений и выбросов.

### Создание копии данных для предобработки

```python
df = autos.copy()
```

### Обработка названий столбцов

### Разделение данных на тренеровочныую и тестовую выборки

Разделим данные на признаки (features) и целевой признак (target). Затем разделим их на обучающую и тестовую выборки.

Так как напрямую стратифицировать непрерывный целевой признак (Price) нельзя, сделаем стратификацию по "корзинам" (bins).

Временно превратим наш непрерывный целевой признак в категориальный, разбив его на группы (корзины, бины), а затем проведем стратификацию по этим группам. После чего удалим временный признак.

```python
# Определяем признаки (X) и целевую переменную (y)
X = df.drop('price', axis=1)
y = df['price']

# Создадим временный столбец с "корзинами" для стратификации
# Используем pd.qcut для создания 10 бинов с равным количеством элементов
# labels=False возвращает числовые метки для каждой корзины
price_bins = pd.qcut(y, q=10, labels=False, duplicates='drop')

# Разделяем данные, используя созданные "корзины" для стратификации
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=TEST_SIZE, 
    random_state=RANDOM_STATE,
    stratify=price_bins
)

# Проверим, что распределения цен в train и test выборках похожи
print("Описательная статистика цен в обучающей выборке:")
display(y_train.describe())
print("\nОписательная статистика цен в тестовой выборке:")
display(y_test.describe())

# Временный столбец price_bins нам больше не нужен, мы его не сохраняли.
# Мы работаем с оригинальными X_train, X_test, y_train, y_test.
```

Объединим X_train и y_train, произведем глубокую предобработку данных только с обучающей выборкой. Перед обучением моделей снова разделим обучающую выборку на X_train и y_train.

```python
train_data = pd.concat([X_train, y_train], axis=1)
```

### Обработка пропусков

Рассмотрим столбцы с пропусками: `vehicle_type`, `gearbox`, `model`, `fuel_type`, `repaired`.

#### `repaired` (Была машина в ремонте или нет)

Посмотрим уникальные значения этого стоблика

```python
# Посчитаем количество уникальных значений столбика "repaired"
train_data['repaired'].value_counts()
```

Предположим, что пропущенные значения в `repaired` означают, что машина не была в ремонте или информация об этом отсутствует. Заполним пропуски значением 'no' или 'unknown'. В данном случае 'no' кажется более логичным, поскольку если бы машина была в ремонте, это, скорее всего, было бы указано.

```python
train_data['repaired'] = train_data['repaired'].fillna('no')
train_data['repaired'].value_counts()
```

#### `vehicle_type`, `gearbox`, `model`, `fuel_type`

Для этих категориальных признаков, где пропуски могут означать отсутствие информации или редкие категории, заполним их значением 'unknown'. Это позволит сохранить эти строки в данных и избежать потери информации, а модели смогут обработать "unknown" как отдельную категорию.

```python
for col in ['vehicle_type', 'gearbox', 'model', 'fuel_type']:
    train_data[col] = train_data[col].fillna('unknown')

train_data.info() # Проверим результат
```

Автомобиль не может быть зарегистрирован позже даты создания объявления или текущей даты. Максимальный год регистрации не может быть больше года скачивания объявления. Минимальный год регистрации обычно принимают, например, 1900.

```python
current_year = train_data['date_crawled'].dt.year.max() # Максимальный год из даты скачивания

# Удаляем строки, где год регистрации выходит за разумные пределы
train_data = train_data[(train_data['registration_year'] >= 1900) & (train_data['registration_year'] <= current_year)]

plt.figure(figsize=(10, 6))
sns.histplot(train_data['registration_year'], bins=50, kde=True)
plt.title('Распределение годов регистрации')
plt.xlabel('Год регистрации')
plt.ylabel('Количество')
plt.show()

train_data['registration_year'].describe()
```

#### `power` (Мощность)

Мощность не может быть 0 л.с. и не может быть нереально высокой (например, 20000 л.с.). Определим разумные границы.

```python
# Установим минимальную разумную мощность, например, 50 л.с. и максимальную 800 л.с.

train_data = train_data[(train_data['power'] >= 50) & (train_data['power'] <= 800)]

plt.figure(figsize=(10, 6))
sns.histplot(train_data['power'], bins=50, kde=True)
plt.title('Распределение мощности')
plt.xlabel('Мощность (л.с.)')
plt.ylabel('Количество')
plt.show()

train_data['power'].describe()
```

## Шаг 4. Корреляционный анализ

Нет смысла искать линейную зависимость между количественными признаками и категориальными. Отсортируем эти признаки друг от друга.

Для начала рассчитаем коэффициенты корреляции между количественными признаками.

### Корреляционный анализ количественных признаков

Ввиду того что не все признаки распределены нормально принимаем решение использовать корреляцию Спирмана.

```python
# Выбираем только количественные признаки
quantitative_features = ['price',
                         'power',
                         'kilometer']
df_quant = train_data[quantitative_features]

# Рассчитываем корреляционную матрицу
corr = df_quant.corr(method='spearman')

# Создаем тепловую карту
plt.figure(figsize=(10, 8))
sns.heatmap(corr, 
            annot=True, 
            fmt=".2f", 
            cmap='cividis',
            vmin=-1, vmax=1,
            square=True)
plt.title('Матрица корреляции Спирмана количественных признаков')
plt.show()
```

**Основные наблюдения:**

1.  **`price` и `power` (корреляция 0.52):**
    *   Это умеренная положительная корреляция.
    *   Она означает, что с увеличением мощности автомобиля (`power`), его цена (`price`) имеет тенденцию увеличиваться. Это вполне логично, так как более мощные автомобили часто позиционируются как более дорогие.

2.  **`price` и `kilometer` (корреляция -0.38):**
    *   Это умеренная отрицательная корреляция.
    *   Показывает, что с увеличением пробега автомобиля (`kilometer`), его цена (`price`) имеет тенденцию снижаться. Это также ожидаемо, поскольку больший пробег обычно ассоциируется с износом и, следовательно, меньшей стоимостью на вторичном рынке.

3.  **`power` и `kilometer` (корреляция 0.12):**
    *   Это очень слабая положительная корреляция.
    *   Значение 0.15 близко к нулю, что говорит о практически отсутствии линейной взаимосвязи между мощностью автомобиля и его пробегом. Другими словами, мощный автомобиль не обязательно имеет высокий или низкий пробег по сравнению с менее мощным.

**Общие выводы:**

*   **Цена (`price`)** является наиболее зависимым признаком от других. Она достаточно хорошо коррелирует с мощностью (положительно) и пробегом (отрицательно), что соответствует здравому смыслу на автомобильном рынке.
*   **Мощность (`power`)** в основном влияет на цену, но слабо связана с пробегом.
*   **Пробег (`kilometer`)** также значительно влияет на цену, но почти не связан с мощностью.

Теперь рассчитаем коэффициенты корреляции между категориальными признаками.

### Корреляционный анализ категориальных признаков

Корреляция Пирсона не подходит для категориальных признаков. Используем коэффициент корреляции Фи, который находит взаимосвязь между категориальными, порядковыми и интервальными признаками. Чем ближе его значение по модулю к 1, тем сильнее связь.

```python
# Определяем интервальные (числовые) признаки для Phi_K
interval_cols = ['price', 'registration_year', 'power']

# Вычисляем матрицу корреляций Phi_K
phik_correlation_matrix = train_data.phik_matrix(interval_cols=interval_cols).round(2)

# Строим тепловую карту Phi_K
plt.figure(figsize=(12, 10)) # Увеличим размер для лучшей читаемости
sns.heatmap(phik_correlation_matrix,
            annot=True,
            cmap='cividis', # Используем ту же цветовую карту для консистентности
            vmin=-1, vmax=1,
            fmt=".2f",
            annot_kws={"size": 9}) # Размер шрифта аннотаций
plt.title("Матрица корреляций Phi_K", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10) # Повернем метки для лучшей читаемости
plt.yticks(rotation=0, fontsize=10)
plt.tight_layout() # Чтобы все поместилось
plt.show()

# Выведем корреляции с целевым признаком 'price'
if 'price' in phik_correlation_matrix.index:
    print("\nКорреляция Phi_K признаков с 'price':")
    print(phik_correlation_matrix['price'].sort_values(ascending=False))
else:
    print("\nСтолбец 'price' не найден в матрице корреляций. Проверьте правильность имени.")
```

    *   **Сильная положительная корреляция.** Это самый сильный фактор, влияющий на цену (после самой цены).
    *   Означает, что конкретная модель автомобиля (`model`) имеет огромное значение для его цены. Различные модели (например, Mercedes S-Class vs. Fiat Punto) имеют существенно разные ценовые категории.
    
    
3.  **`registration_year` (0.47):**
    *   **Средняя положительная корреляция.**
    *   Чем новее год регистрации (`registration_year`), тем выше цена (`price`). Это логично: более новые автомобили стоят дороже из-за меньшего износа, актуальности технологий и меньшего пробега.


4.  **`power` (0.46):**

5.  **`brand` (0.36):**

7.  **`repaired` (0.30):**
    *   **Умеренная корреляция.** Точное направление (положительная или отрицательная) не всегда очевидно из одной только величины Phi_K для категориальных признаков без дальнейшего анализа, но часто для `repaired` (был ли ремонт) ожидается отрицательная связь: отремонтированные автомобили могут стоить дешевле. Однако Phi_K может улавливать и более сложные зависимости.

**Анализ мультиколлинеарности и её устранение**

**Мультиколлинеарность** — это явление, при котором два или более признака в наборе данных сильно коррелируют друг с другом. Это может быть проблемой, особенно для линейных моделей, так как усложняет интерпретацию вклада каждого отдельного признака в конечный результат и может сделать модель нестабильной.

**Выявление мультиколлинеарности в данных:**

В нашей матрице Phi_K мы видим несколько пар с очень высокой корреляцией:

1.  **`model` и `brand` (корреляция 1.00):**
    *   **Причина:** Это идеальная мультиколлинеарность. Каждая модель (`model`) однозначно принадлежит одному бренду (`brand`). Например, модель "Golf" всегда будет принадлежать бренду "Volkswagen". Информация в признаке `brand` полностью дублируется информацией, содержащейся в признаке `model`.
    *   **Решение:** Необходимо удалить один из этих признаков.


2.  **`model` и `vehicle_type` (корреляция 0.90):**
    *   **Причина:** Очень высокая корреляция. Конкретная модель почти всегда имеет один и тот же тип кузова. Например, "Ford Focus" в большинстве случаев будет хэтчбеком или седаном. Информация в `vehicle_type` в значительной степени дублируется признаком `model`.
    *   **Решение:** Также рекомендуется удалить один из этих признаков.

**Оставляем тот признак, который сильнее коррелирует с целевой переменной (`price`)**.

*   Сравниваем `model` (корреляция с `price` = **0.57**) и `brand` (корреляция с `price` = 0.36).
    *   **Вывод:** `model` является более сильным предиктором. **Удаляем `brand`**.

*   Сравниваем `model` (корреляция с `price` = **0.57**) и `vehicle_type` (корреляция с `price` = 0.27).
    *   **Вывод:** `model` снова является гораздо более сильным предиктором. **Удаляем `vehicle_type`**.

**Важное примечание о моделях:**
*   Для **линейных моделей** (например, Линейная регрессия) устранение мультиколлинеарности является **критически важным**.
*   Для **древовидных моделей** (Random Forest, Gradient Boosting, LightGBM) мультиколлинеарность является **менее серьезной проблемой**, так как алгоритм построения деревьев сам выберет наиболее информативный признак на каждом шаге. Однако удаление избыточных признаков все равно является хорошей практикой, так как это упрощает модель и может незначительно ускорить обучение.

```python
X_train = train_data.drop('price', axis=1)
y_train = train_data['price']
```

### Кодирование признаков и построение пайплайнов

Для корректной работы моделей необходимо закодировать категориальные признаки. LinearRegression требует OneHotEncoder, в то время как CatBoost и LightGBM могут работать с категориальными признаками напрямую, что эффективнее. Мы создадим два подхода:

1. **Пайплайн с OneHotEncoder** для `LinearRegression` и `LightGBM`.

2. **Прямая передача категориальных признаков** для `CatBoost`.

```python
# Определяем категориальные и числовые признаки
categorical_features = ['gearbox', 'model', 'fuel_type', 'repaired']
numerical_features = ['registration_year', 'power', 'kilometer']

# Создаем пайплайн для предобработки данных
# Числовые признаки масштабируем, категориальные - кодируем OHE
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough'
)
```

### Обучение и оценка моделей

Теперь обучим три модели: `LinearRegression` как базовую, и две модели градиентного бустинга — `CatBoostRegressor` и `LightGBM`. Для каждой замерим время обучения, время предсказания и RMSE.

#### Модель 1: LinearRegression

```python
# Создаем полный пайплайн с моделью
pipeline_lr = Pipeline(steps=[('preprocessor', preprocessor),
                              ('regressor', LinearRegression())])

# Запускаем кросс-валидацию и замеряем метрики
print("Запускаем кросс-валидацию для LinearRegression...")

# Используем cv=3, чтобы быть последовательными с другими моделями
# n_jobs=-1 использует все ядра процессора для ускорения
cv_results_lr = cross_validate(
    estimator=pipeline_lr,
    X=X_train,
    y=y_train,
    cv=2,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1,
    return_train_score=False # Нам не нужна оценка на train
)

# Извлекаем и выводим результаты
# Результаты хранятся в словаре cv_results_lr. Мы их усредняем.

# Средний RMSE на валидационных фолдах
# (берем с минусом, так как scoring='neg_..._error')
rmse_lr_cv = -np.mean(cv_results_lr['test_score'])

# Среднее время обучения модели на каждом фолде
fit_time_lr_cv = np.mean(cv_results_lr['fit_time'])

# Среднее время предсказания на каждом фолде
score_time_lr_cv = np.mean(cv_results_lr['score_time'])

print("\n--- Результаты для LinearRegression (на кросс-валидации) ---")
print(f"Средний RMSE: {rmse_lr_cv:.2f}")
print(f"Среднее время обучения (fit_time): {fit_time_lr_cv:.4f} секунд")
print(f"Среднее время предсказания (score_time): {score_time_lr_cv:.4f} секунд")
```

#### Модель 2: CatBoostRegressor

`CatBoost` лучше всего работает, когда мы передаем ему категориальные признаки напрямую, без OHE.

```python
# Определяем модель
# Мы будем использовать CatBoost с базовыми, но адекватными параметрами для сравнения.
# verbose=0 отключает вывод логов во время кросс-валидации, чтобы не засорять вывод.
model_catboost = CatBoostRegressor(
    random_state=RANDOM_STATE,
    cat_features=categorical_features,
    iterations=300,        # Разумное количество для быстрой, но качественной оценки
    learning_rate=0.2,
    verbose=0              # Отключаем логирование для чистоты вывода
)

# Запускаем кросс-валидацию и замеряем метрики
print("Запускаем кросс-валидацию для CatBoostRegressor...")

# Используем те же параметры (cv=2, n_jobs=-1), что и для LinearRegression
cv_results_cat = cross_validate(
    estimator=model_catboost,
    X=X_train,
    y=y_train,
    cv=2,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1,
    return_train_score=False
)

# Средний RMSE на валидационных фолдах
rmse_cat_cv = -np.mean(cv_results_cat['test_score'])

# Среднее время обучения модели на каждом фолде
fit_time_cat_cv = np.mean(cv_results_cat['fit_time'])

# Среднее время предсказания на каждом фолде
score_time_cat_cv = np.mean(cv_results_cat['score_time'])

print("\n--- Результаты для CatBoostRegressor (на кросс-валидации) ---")
print(f"Средний RMSE: {rmse_cat_cv:.2f}")
print(f"Среднее время обучения (fit_time): {fit_time_cat_cv:.4f} секунд")
print(f"Среднее время предсказания (score_time): {score_time_cat_cv:.4f} секунд")
```

#### Модель 3: LightGBM (с подбором гиперпараметров)

Сосредоточимся на поиске лучших гиперпараметров для модели — `LGBMRegressor`. Мы будем использовать `RandomizedSearchCV` для эффективного поиска в пространстве параметров, оценивая каждую комбинацию с помощью 2-х фолдовой кросс-валидации.

```python
# Определяем пайплайн для LightGBM
# Определяем списки признаков (если они не были определены ранее в ноутбуке)
categorical_features = ['gearbox', 'model', 'fuel_type', 'repaired']
numerical_features = ['registration_year', 'power', 'kilometer']

# Создаем препроцессор, который будет шагом в нашем пайплайне
# Он масштабирует числовые признаки и кодирует категориальные
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

# Создаем полный пайплайн: сначала предобработка, затем модель
pipeline_lgbm = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', lgb.LGBMRegressor(random_state=RANDOM_STATE))
])


# Определяем сетку параметров и запускаем поиск
# Ультрапростая сетка для максимально быстрого поиска
param_dist_ultrafast = {
    'regressor__n_estimators': [100, 150],
    'regressor__learning_rate': [0.2, 0.3],
    'regressor__max_depth': [5, 7],
    'regressor__num_leaves': [31, 41],
}

print("Подбираем параметры и оцениваем LGBMRegressor...")

# Инициализируем RandomizedSearchCV
# n_iter=6 и cv=2
random_search_lgbm = RandomizedSearchCV(
    estimator=pipeline_lgbm,
    param_distributions=param_dist_ultrafast,
    n_iter=6,
    cv=2,
    scoring='neg_root_mean_squared_error',
    random_state=RANDOM_STATE,
    n_jobs=-1,
    verbose=1
)

# Запускаем поиск на обучающих данных
random_search_lgbm.fit(X_train, y_train)


# Извлекаем и выводим результаты
print("\nПодбор параметров для LGBM завершен.")

# Находим индекс лучшей модели в результатах поиска
best_index = random_search_lgbm.best_index_

# Извлекаем среднее время обучения для лучшей модели из результатов кросс-валидации
fit_time_lgbm_cv = random_search_lgbm.cv_results_['mean_fit_time'][best_index]

# Извлекаем среднее время предсказания (оценки)
score_time_lgbm_cv = random_search_lgbm.cv_results_['mean_score_time'][best_index]

# Извлекаем лучший RMSE (не забываем убрать минус)
rmse_lgbm_cv = -random_search_lgbm.best_score_

# Сохраняем лучшую модель (весь пайплайн) для финального тестирования.
# Она уже обучена на всей X_train, y_train.
best_lgbm_model = random_search_lgbm.best_estimator_

# Выводим результаты в едином формате
print("\n--- Результаты для LGBMRegressor (tuned, на кросс-валидации) ---")
print(f"Лучшие параметры: {random_search_lgbm.best_params_}")
print(f"Средний RMSE: {rmse_lgbm_cv:.2f}")
print(f"Среднее время обучения (fit_time): {fit_time_lgbm_cv:.4f} секунд")
print(f"Среднее время предсказания (score_time): {score_time_lgbm_cv:.4f} секунд")
```

### Анализ моделей

Соберем все результаты в одну таблицу.

```python
# Создаем словарь с результатами 
results_data = {
    'Model': [
        'LinearRegression',
        'CatBoostRegressor (default)',
        'LGBMRegressor (tuned)'
    ],

    'CV RMSE': [
        rmse_lr_cv,
        rmse_cat_cv,
        rmse_lgbm_cv
    ],

    'Mean Fit Time (s)': [
        fit_time_lr_cv,
        fit_time_cat_cv,
        fit_time_lgbm_cv
    ],

    'Mean Predict Time (s)': [
        score_time_lr_cv,
        score_time_cat_cv,
        score_time_lgbm_cv
    ]
}

# Создаем и выводим итоговую таблицу
# Преобразуем словарь в DataFrame
results_df = pd.DataFrame(results_data)

print("Сравнительная таблица моделей (оценка на кросс-валидации):")

display(results_df.style.hide_index())

# Делаем вывод на основе таблицы

print("\n--- Анализ результатов ---")

print("1. Качество предсказания (CV RMSE):")
print("   - LinearRegression показывает значительно худшее качество и не подходит для решения задачи.")
print("   - Модели градиентного бустинга (CatBoost и LGBM) демонстрируют отличное и очень близкое качество, с небольшим преимуществом у настроенной модели LGBM.")

print("\n2. Время обучения (Mean Fit Time):")
print("   - LGBMRegressor является абсолютным лидером по скорости обучения. Он обучается в несколько раз быстрее, чем CatBoost.")
print("   - LinearRegression также обучается очень быстро, но ее качество неприемлемо.")

print("\n3. Время предсказания (Mean Predict Time):")
print("   - Все модели показывают очень высокую скорость предсказания (доли секунды), что удовлетворяет требованиям заказчика.")

print("\n--- Вывод ---")
# Автоматически находим название лучшей модели по минимальному RMSE
best_model_name = results_df.loc[results_df['CV RMSE'].idxmin()]['Model']

print(f"Лучшей моделью для решения поставленной задачи является '{best_model_name}'.")
print("Она обеспечивает наилучший баланс между высочайшим качеством предсказания (самый низкий RMSE) и превосходной скоростью обучения.")
```

**Вывод:**
Модель **`LGBMRegressor (tuned)`** является оптимальным выбором. Она предлагает наилучший баланс между высочайшим качеством предсказания и превосходной скоростью обучения.

### Финальное тестирование

Проверим итоговые показатели выбранной модели на тестовой выборке.

```python
# Подготовка тестовой выборки
# Мы должны применить к X_test те же преобразования, что и к X_train.
# Главное - удалить те же самые столбцы.
# Создаем копию, чтобы не изменять оригинальный X_test
X_test_processed = X_test.copy()

# Заполняем пропуски так же, как в train (хотя в удаляемых колонках это не так важно)
# ВАЖНО: Мы не удаляем дубликаты и не фильтруем строки из тестовой выборки!
# Тестовая выборка должна оставаться "нетронутой", чтобы имитировать реальные данные.
# Мы только приводим ее к тому же формату столбцов, что и обучающую.
X_test_processed['repaired'] = X_test_processed['repaired'].fillna('no')
for col in ['vehicle_type', 'gearbox', 'model', 'fuel_type']:
    X_test_processed[col] = X_test_processed[col].fillna('unknown')

# Определяем список столбцов для удаления (тот же, что и для train)
cols_to_drop = [
    'number_of_pictures', 'date_crawled', 'date_created', 'last_seen',
    'postal_code', 'registration_month', 'brand', 'vehicle_type'
]

# Удаляем столбцы из подготовленной тестовой выборки
X_test_processed.drop(cols_to_drop, axis=1, inplace=True)

# Используем лучшую модель и делаем предсказание
best_model = best_lgbm_model

print("\n--- Финальное тестирование лучшей модели: LGBMRegressor (tuned) ---")
print("Делаем предсказания на обработанной тестовой выборке...")

start_pred_time = time.time()
# ИСПОЛЬЗУЕМ ОБРАБОТАННУЮ ВЫБОРКУ X_test_processed
final_predictions = best_model.predict(X_test_processed)
end_pred_time = time.time()

# Рассчитываем и выводим финальные метрики
final_prediction_time = end_pred_time - start_pred_time
final_rmse = mean_squared_error(y_test, final_predictions, squared=False)

print("\n--- Итоги финального теста ---")
print(f"Финальный RMSE на тестовой выборке: {final_rmse:.2f}")
print(f"Время предсказания для тестовой выборки: {final_prediction_time:.4f} секунд")

if final_rmse < 2500:
    print(f"Требование заказчика (RMSE < 2500) ВЫПОЛНЕНО.")
else:
    print(f"Требование заказчика (RMSE < 2500) НЕ ВЫПОЛНЕНО.")
```
